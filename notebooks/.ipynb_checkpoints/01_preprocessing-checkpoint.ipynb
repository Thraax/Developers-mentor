## Constants

```python
DATA_PATH = 'C:/Users/Thrax/Stackover_survey/Survey_analysis/data/raw/survey_results_public.csv'
EXPORT_PATH = 'C:/Users/Thrax/Stackover_survey/Survey_analysis/data/processed/preprocessed_df.pkl'


REPLACE_DICT = {
    'YearsCodePro': {'Less than 1 year': 0, 'More than 50 years': 51}, 
    'YearsCode':    {'Less than 1 year': 0, 'More than 50 years': 51}, 
    'Age1stCode':   {'Older than 85':86,    'Younger than 5 years':4}}
```

## Import tools

```python
import pandas as pd
from pandas.api.types import is_string_dtype
import numpy as np
import pickle
```

## Functions

```python
def split_answers(pd_series, delimiter):
    
    # Sub-function that check if the series splitable
    def is_splitable(pd_series, delimiter):
        return pd_series.str.contains(delimiter)
    
    # Sub-function that split single answer
    def split_answer(pd_series, delimiter):
        return pd_series.str.split(delimiter)
    
    # Check if there multiple answers in the series, if not return original
    splitable_values = is_splitable(pd_series, delimiter)
    
    # use any() method to check if there any false in the series
    if not splitable_values.any():
        return pd_series # if there is no multiple answers, return the original
   
    # What if there multiple answers? aka: 'else'
    
    splited_series = split_answer(pd_series, delimiter)
    
    #Replace the NaNs to avoid any error
    null_mask = splited_series.isnull()
    splited_series.loc[null_mask] = splited_series.loc[null_mask].apply(lambda x: [])
    
    return splited_series
    
```

```python
raw_df = pd.read_csv(DATA_PATH) # Make a copy from the original data
df_copy = raw_df.copy()
```

## Values replacment

```python
for col, replacment in REPLACE_DICT.items():
    df_copy[col] = df_copy[col].replace(replacment).astype(np.float64)
```

## Split multiple answers

```python
categorical_cols = df_copy.select_dtypes(include='object').columns.tolist()

for col in categorical_cols:
    df_copy[col] = split_answers(df_copy[col], ';')
```
